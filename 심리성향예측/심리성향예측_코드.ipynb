```python
drop_list = ['familysize', 'hand', 'engnat', 'gender']
```

```python
label = LabelEncoder()
train['race'] = label.fit_transform(train['race'])
train['religion'] = label.fit_transform(train['religion'])
train['age_group'] = label.fit_transform(train['age_group'])
train = train.drop(drop_list, axis=1, inplace = False)
train.head()
```

```python
label = LabelEncoder()
test['race'] = label.fit_transform(test['race'])
test['religion'] = label.fit_transform(test['religion'])
test['age_group'] = label.fit_transform(test['age_group'])
test = test.drop(drop_list, axis=1, inplace = False)
test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train = train.drop(["voted"], axis=1)
Y_train = train["voted"]
X_test  = test.copy()
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)
```
